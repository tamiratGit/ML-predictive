```python
df_data = data.copy()#copy the database
```

Note- column 13 has 376 missing values which out of 452 rows which make more tha 60% of the data, it has no significant for the test, need to be drop.

```python
df_data= df_data.drop(columns=[13])#drop column 13
```

```python
#fill nan cells of the above columns except 13 by the median of the column
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

X_df = df_data.iloc[:,:-1] #select all features except target feature
X = np.array(X_df) #convert it to array (Simple Imputer doesn't work with dataframes)
imp = SimpleImputer(missing_values=np.nan, strategy='median') #create imp object to impute median in all missing values 
imp = imp.fit(X) #calculate median values of the features with missing values
X_imp = imp.transform(X) #fill dataset with median values wherever finds missing values
X_imp.shape
```

```python
np.isnan(X_imp).sum().sum() #total number of missing values in the dataset
```

```python
y= df_data['target']
X_train, X_test, y_train, y_test = train_test_split(X_imp, y, random_state=0, test_size=.2)
X_train.shape, y_test.shape
```

```python
#print("Test Score with only selected features: {:.3f}".format(clf.score(X_test_selected, y_test)))
#print("Train Score with all features: {:.3f}".format(clf.score(X_train_selected, y_train)))
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import chi2

output = pd.DataFrame(index=None, columns=['kn','train_score', 'test_score'])

train_score=[]
test_score=[]

clf = KNeighborsClassifier(n_neighbors=3)
def uni_vari(X_train,X_test, y_train, y_test,kn):
    
    uni_model = SelectKBest(score_func=f_regression, k=kn)
    uni_model.fit(X_train, y_train)

    # transform training set
    X_train_selected = uni_model.transform(X_train)
    X_test_selected = uni_model.transform(X_test)


    clf.fit(X_train, y_train)

    clf.fit(X_train_selected, y_train)

    test_score=clf.score(X_test_selected, y_test)
    train_score=clf.score(X_train_selected, y_train)
    #output=output.append(pd.Series({'kn':n,'train_score':train_score, 'test_score':test_score}),ignore_index=True )
    return test_score, train_score 
        
```

```python
number_of_features= [20, 50, 100, 150, 200]
for n in number_of_features:
    test_score,train_score=uni_vari(X_train,X_test, y_train, y_test,n)
    output= output.append(pd.Series({'kn':n,'train_score':train_score, 'test_score':test_score}),ignore_index=True )
output
```

Model-Based Feature Selection

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
knn_clf = KNeighborsClassifier()

def model_fun(X_train,X_test, y_train, y_test,kn):
    param_grid = {'n_neighbors':[3],
                 }

    grid_search = GridSearchCV(knn_clf , param_grid,scoring = 'accuracy',return_train_score=True,cv=2)
    grid_search.fit(X_train, y_train)
    SelectFromModel(grid_search,  max_features=kn)
    
    print('test')
    print(grid_search.cv_results_['mean_test_score'])
    print('train')
    print(grid_search.cv_results_['mean_train_score'])
```

```python
number_of_features= [20, 50, 100, 150, 200]
for n in number_of_features:
    model_fun(X_train,X_test, y_train, y_test,n)
```

Iterative Feature Selection

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
output3 = pd.DataFrame(index=None, columns=['kn','train_score', 'test_score'])
def iter_function(X_train,X_test, y_train, y_test,kn):
    select_REF = RFE(RandomForestClassifier(n_estimators=100, random_state=42),
    n_features_to_select=kn)
    select_REF.fit(X_train, y_train)

    X_train_rfe= select_REF.transform(X_train)
    X_test_rfe= select_REF.transform(X_test)

    score = LogisticRegression().fit(X_train_rfe, y_train).score(X_test_rfe, y_test)
    #print("Test score: {:.3f}".format(score))
    score_t = LogisticRegression().fit(X_train_rfe, y_train).score(X_train_rfe, y_train)
    #print("Train score: {:.3f}".format(score_t))
    return score,score_t
   
```

```python
number_of_features= [20, 50, 100, 150, 200]
for n in number_of_features:
    test_score,train_score=iter_function(X_train,X_test, y_train, y_test,n)
    output3= output3.append(pd.Series({'kn':n,'train_score':train_score, 'test_score':test_score}),ignore_index=True )
#print(output2)
output3
```
